10. Adicione os seguintes rótulos ao nosso gráfico:

* Título: "Quantidade de Ingressos x Gorjetas"
* Eixo x: "Valor do Bilhete"
* eixo y: "Dicas"

```python
p={'Female':'blue','Male':'green'}

plt.figure(figsize=(20,5))
sns.scatterplot(x='total_bill', y = 'tip',data = df, hue = 'sex', style = 'smoker', 
                s = 200, alpha = 0.5, palette = p, legend="full")
plt.xlabel('Importe del Tickets')
plt.ylabel('Propina')
plt.title('Importe del Tickets vs Propinas',fontsize=20)
```

# Distplot

1. Gere um histograma para a variável total_bill


```python
plt.figure(figsize=(20,5))

sns.distplot(df.total_bill, kde=True, hist=True, bins=15)
plt.xlabel('Total bill')
plt.ylabel('Frecuencia')
plt.title('Distribución de total_bill')

plt.show()
```

2. Compare no mesmo gráfico as distribuições de total_bill e tip


```python
sns.set(style="darkgrid")

fig = plt.figure(figsize=(20,5))

fig = sns.kdeplot(df['total_bill'], shade = True, color="r")
fig = sns.kdeplot(df['tip'], shade = True, color="b")
plt.show()
```

# Boxplot

1. Gere um boxplot onde x = dia e y = total_bill, atribua as seguintes cores:

# Violin Plot

# Scatterplot com Plotly

1. Gere um gráfico de dispersão em que o eixo x é "total_bill" e o eixo y é "tip"

```python
fig = px.scatter(df, x = "total_bill", y = "tip")
fig.show()
```

2. Atribua as cores com base na variável "day"

```python
fig = px.scatter(df, x="total_bill", y="tip",color='day')
fig.show()
```

3. Modifique o tamanho dos pontos com base na variável "size"

```python
fig = px.scatter(df, x="total_bill", y="tip",color='day',size='size')
fig.show()
```

4. Gere um gráfico de dispersão 3d onde a variável **"x"** é "total_bill", a variável **"y"** é "ponta" e a variável **"z"** é "tamanho"

```python
fig = px.scatter_3d(df, x="total_bill", y="tip",z='size')
fig.show()
```

5. Modificar el grafico anterior y asignar un color en base a la variable "day"


```python
fig = px.scatter_3d(df, x="day", y="tip",z='total_bill',color='day')
fig.show()
```

# Violin Plot

Gere um gráfico de caixa em que o eixo x é "fumante" e o eixo y é "total_bill"

# Pie Plot
